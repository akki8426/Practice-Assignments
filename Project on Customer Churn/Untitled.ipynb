# Customer Churn Analysis

Customer Churn Is When a Company's Customers Stop Doing Bussiness With That Company.Bussinesses Are Very Keen On Measuring Churn Because Keeping An Existing Customer Is Far Expensive Than Acquiring a New Customer. New Bussiness Involves Working Leads Through a Sales funnel, Using Marketing and Sales Budgets To Gain Additional Customers. Existing Customers Will Often Have a Higher Volume Of Services Consumption and Can Generate Additional Customer Refferal

Customer Retention Can Be Achived With Good Customer Service And Products. But The Most Effective Way For a Company To Prevent Atteration Of Customers Is To Truly Know Them.
The Vast Volumes Of Data Collected About Customers Can Be Used To Built Churn Prediction Models. Knowing Who is Mostly Likely To Defect Means That Company Can Priorities Focused Marketing Efforts On That Subsets Of Theire Customer Base.

Preventing Customer Churns Is Critically Important To be Telecommunication Sectors, As The Barriers To Entry For Swithing Services Are So Low.

You Can Examine Customer Data From IBM Sample Data Sets With The Aim Of Building And Compairing Several Customer Churn Prediction Models.

The Data Set Includes Information About:
    
    Customers Who Left Within The Last Month - The Columns Are Called Churn Services That Each Customer Has Signed Up For - Phones, Multiple Lines, Internet, Online Security, Online BackUp, Device Protection, Tech Support and Streaming TV and Movies.
    Customer Account Information - How Long They've Been Customer, Contract, Payment Method, Paperless Billing, Monthly Carges and Total charges.
    Demographic Info About Customers - Gender, Age Range and If They Have Partners And Dependents.

Dataset link : __(https://github.com/dsrscientist/DSData/blob/master/Telecom_customer_churn.csv)__

### Machine Learning Model for classification With Evaluation Metrics

````python
def classify (clf,xa,y):
    x_train,x_test,y_train,y_test=train_test_split(xa,y,test_size=0.25,random_state=868)
    
    #Training the model:
    clf.fit(x_train,y_train)
    
    #Predicting y-test:
    pred=clf.predict(x_test)
    
    #Accuracy Score:
    acc_score=(accuracy_score(y_test,pred))
    print("The Accuracy Score is :",acc_score)
    
    #Confusion Matrix:
    print("confusion matric :\n",confusion_matrix(y_test,pred))
    
    #Classification Report:
    print("classification report :\n",classification_report(y_test,pred))
    
    #Cross validation Score:
    cv_score=(cross_val_score(clf,xa,y,cv=5).mean())
    print("Cross Validation Score :",cv_score)
    
    #Result of Accuracy:
    result=acc_score-cv_score
    print("Accuracy Score-Cross validation Score :",result)
````

### Logistic Regression

````python
clf=LogisticRegression()
classify(clf,xa,y)
````

### Decision Tree Classifier

````python
clf=DecisionTreeClassifier()
classify(clf,xa,y)
````
